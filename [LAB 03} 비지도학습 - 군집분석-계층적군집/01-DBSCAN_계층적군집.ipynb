- 두 점 사이의 거리를 어떻게 측정할건지 정해야함 (직선거리 - 유클리디안 / 블록거리 - 맨하탄)

## #02 클러스터 접근 절차의 체크 포인트
### [1] 거리 측정 방식 (metric)
- 서로 가까운 점끼리 합치는 방식이라서 두 점 사이의 거리를 구하는 방식이 중요함


#### (1) 유클리디안 거리 측정
- sklearn 기본값
- 두 점 사이의 거리를 계싼할 떄 많이 활용
- 두 점 상의 직선 상의 거리
  

$
d(p, q) = \sqrt{(q_1 - p_1)^2 + (q_2 - p_2)^2 + \cdots + (q_n - p_n)^2}
$

|값|설명|
|-|-|
|d(p,q)|두 점 p와 q 사이의 유클리디안 거리|
|q1,q2,q3|각각 점 p와 q 의 좌표|



#### (2) 멘하탄 거리 측정
- 두 점 사이의 거리를 계싼할 떄 각 좌표 축을 따라 이동한 거리의 총합
- 두 점 간의 좌표의 차의 절대값으로 계산
  
$
d(p, q) = |q_1 - p_1| + |q_2 - p_2| + \cdots + |q_n - p_n|
$

|값|설명|
|-|-|
|d(p,q)|두 점 p와 q 사이의 멘하탄 거리|
|q1,q2,q3|각각 점 p와 q 의 좌표|



### [3] 데이터 결합 방식 (linkage)
- 데이터를 병합하는 형태가 군집 결과에 영향을 미치므로 이 과정 역시 중요함
- 데이터를 병합하는 과정은 덴드로그램 형태로 나타남

#### [1] Min(Single Link)
- 군집과 군집의 거리를 구할 때 가장 최소거리인 것을 유사도로 측정하는 방식
- 가까운 점이 하나라도 있으면 병합되면서 사슬 현상이 발생할 수 있음
- 비선형 군집에는 유리하지만 노이즈에 민감하고, 길게 늘어진 군집이 생성될 수 있음

#### [2] Max (Complete Link)
- 군집과 군집의 거리를 구할 때 가장 최대 거리인 것을 유사도로 측정하는 방식
- 군집 간 거리를 가장 먼 두점으로 정의하기 떄문에 한 점만 가까우면 되는 Min 과는 달리 전반적으로 모든 점이 가까워야 병합되고, 조밀한 군집이 만들어져



#### [3] Average Link
- 군집과 군집의 거리를 구할 떄 거리의 평균을 구해 유사도로 측정하는 방식
- Min 과 Max 의 절충형, 노이즈,사슬 문제가 완화될 수 있음

#### [4] Centroids
- sklearn 은 이 방식을 지원하지 않음
- 군집과 군집의 거리를 구할 때 데이터의 중심점 거리를 유사도로 측정하는 방식

#### [5] Ward
- 두 개의 군집이 병합되었을 때 증가하는 변동성(군집 내 제곱합, 분산) 의 양으로 유사도를 측정하는 방식
- 두 군집을 병합했을 떄, 군집 내 제곱합 (분산) 이 가장 적게 증가하도록 병합


#### 변동성
- 두 군집의 중앙값과 두 군집에 있는 모든 데이터와의 거리의 합에서 군집 간에 형성되는 거리를 뺸 값
- 결국은 병합되는 클러스터의 분산을 최소화 한다는 의미



## #03. 군집 분석 비교

### [1] 군집 분석 종류별 특징 비교
| 구분 | KMeans | DBSCAN | Agglomerative Clustering |
|---|---|---|---|
| 기본 아이디어 | 중심점 기준 분할 | 밀도 기반 연결 | 거리 기반 병합 |
| 군집 수 지정 | 사전 지정 필요 (k) | ❌ 불필요 | 선택 가능 (n_clusters 또는 거리 기준) |
| 이상치 처리 | ❌ 없음 | ⭕ 자동 감지 | ❌ 없음 |
| 군집 형태 | 구형(원형) 가정 | 자유로운 형태 | 비교적 자유 |
| 거리 기준 | 유클리드 중심 거리 | 이웃 밀도 | 군집 간 거리 |
| 결과 안정성 | 초기값 영향 큼 | 파라미터 민감 | ⭐ 결정론적 (항상 동일) |
| 확장성 | ⭕ 대용량에 매우 유리 | 중간 | ❌ 대용량에 불리 |
| 해석 용이성 | 높음 | 중간 | ⭐ 덴드로그램으로 매우 높음 |
| 대표 사용 목적 | 빠른 분할·세분화 | 이상치 탐지 | 구조 이해·계층 해석 |
